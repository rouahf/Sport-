```python
import cv2
import mediapipe as mp
import numpy as np

# Initialisation des variables de détection
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# Variables pour le compteur de curl
counter = 0 
stage = None

# Setup pour la détection des gestes des mains avec Mediapipe
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

# Fonction pour calculer l'angle
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180:
        angle = 360 - angle
    return angle

# Setup caméra
cap = cv2.VideoCapture(0)

# Fonction pour calculer la distance entre deux points
def calculate_distance(p1, p2):
    return np.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            continue

        # Recolorer l'image pour MediaPipe
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Faire la détection
        results = pose.process(image)

        # Recolorer l'image pour OpenCV
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extraire les repères
        try:
            landmarks = results.pose_landmarks.landmark
            if results.pose_landmarks:
                # Obtenir les coordonnées des mains
                hand_results = hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

                if hand_results.multi_hand_landmarks:
                    for hand_landmarks in hand_results.multi_hand_landmarks:
                        # Dessiner les repères de la main
                        mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

                        # Obtenir les positions du pouce et de l'index
                        thumb_tip = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP]
                        index_tip = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]

                        # Calculer la distance entre le pouce et l'index
                        distance = calculate_distance([thumb_tip.x, thumb_tip.y], [index_tip.x, index_tip.y])

                        # Détecter le geste de Finger Snapping
                        if distance < 0.05:  # Distance seuil pour le geste de pincement
                            cv2.putText(frame, 'Finger Snapping', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                        
            # Récupérer les coordonnées des articulations pour le comptage des curls
            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]

            # Calculer l'angle
            angle = calculate_angle(shoulder, elbow, wrist)

            # Visualiser l'angle
            cv2.putText(frame, str(angle), tuple(np.multiply(elbow, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

            # Logique du compteur de curls
            if angle > 160:
                stage = "down"
            if angle < 30 and stage == 'down':
                stage = "up"
                counter += 1
                print(counter)

        except:
            pass

        # Affichage des données du compteur
        cv2.rectangle(frame, (0, 0), (225, 73), (245, 117, 16), -1)

        # Afficher le nombre de répétitions
        cv2.putText(frame, 'REPS', (15, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(frame, str(counter),
                    (10, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

        # Afficher l'état
        cv2.putText(frame, 'STAGE', (65, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(frame, stage,
                    (60, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

        # Dessiner les repères du corps
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))

        # Affichage du flux vidéo avec les détections
        cv2.imshow('Mediapipe Feed', frame)

        # Quitter si 'q' est pressé
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np

# Initialisation des variables de détection
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# Variables pour le compteur de curl
counter = 0 
stage = None

# Setup pour la détection des gestes des mains avec Mediapipe
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

# Fonction pour calculer l'angle
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180:
        angle = 360 - angle
    return angle

# Setup caméra
cap = cv2.VideoCapture(0)

# Fonction pour calculer la distance entre deux points
def calculate_distance(p1, p2):
    return np.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            continue

        # Recolorer l'image pour MediaPipe
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Faire la détection
        results = pose.process(image)

        # Recolorer l'image pour OpenCV
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extraire les repères
        try:
            landmarks = results.pose_landmarks.landmark
            if results.pose_landmarks:
                # Obtenir les coordonnées des mains
                hand_results = hands.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

                if hand_results.multi_hand_landmarks:
                    for hand_landmarks in hand_results.multi_hand_landmarks:
                        # Dessiner les repères de la main
                        mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

                        # Obtenir les positions du pouce et de l'index
                        thumb_tip = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_TIP]
                        index_tip = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]

                        # Calculer la distance entre le pouce et l'index pour le Finger Snapping
                        distance = calculate_distance([thumb_tip.x, thumb_tip.y], [index_tip.x, index_tip.y])

                        # Détecter le geste de Finger Snapping
                        if distance < 0.05:  # Distance seuil pour le geste de pincement
                            # Positionner le texte "Finger Snapping" à droite de l'écran
                            text_x = frame.shape[1] - 200  # Décalage de 200 pixels à partir du bord droit
                            cv2.putText(frame, 'Claquement de doigts ', (text_x, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

                        # Détecter le geste de Thumb Up
                        thumb_base = hand_landmarks.landmark[mp_hands.HandLandmark.THUMB_CMC]  # Base du pouce
                        # Si le pouce est élevé (comparaison de la position du pouce par rapport à l'index et au reste des doigts)
                        if thumb_tip.y < index_tip.y and thumb_base.y < index_tip.y:
                            # Positionner le texte "Thumb Up" à droite de l'écran
                            cv2.putText(frame, 'pouce levé', (text_x, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                        
            # Récupérer les coordonnées des articulations pour le comptage des curls
            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]

            # Calculer l'angle
            angle = calculate_angle(shoulder, elbow, wrist)

            # Visualiser l'angle
            cv2.putText(frame, str(angle), tuple(np.multiply(elbow, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

            # Logique du compteur de curls
            if angle > 160:
                stage = "down"
            if angle < 30 and stage == 'down':
                stage = "up"
                counter += 1
                print(counter)

        except:
            pass

        # Affichage des données du compteur
        cv2.rectangle(frame, (0, 0), (225, 73), (245, 117, 16), -1)

        # Afficher le nombre de répétitions
        cv2.putText(frame, 'REPS', (15, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(frame, str(counter),
                    (10, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

        # Afficher l'état
        cv2.putText(frame, 'STAGE', (65, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(frame, stage,
                    (60, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)

        # Dessiner les repères du corps
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))

        # Affichage du flux vidéo avec les détections
        cv2.imshow('Mediapipe Feed', frame)

        # Quitter si 'q' est pressé
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
